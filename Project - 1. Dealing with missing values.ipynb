```python
import pandas as pd

#taking csv file
fifa_data=pd.read_csv('E:\DBMI\Mini Project\FIFA18CompletePlayerDataset.csv', index_col=0)
fifa_data.describe()
```

```python
#finding size of dataset
fifa_data.shape
```

```python
#columns present in our dataset
fifa_data.columns
```

```python
#finding if any value is missing
fifa_data.isna().values.any()
```

```python
#answer came as True, count of missing values
fifa_data.isna().sum().sum()
```

```python
#copying actual file in new file
data=fifa_data.copy()
data.head(10)
```

```python
#checking for missing value columns
data.isna().sum()
```

```python
#checking for missing value in string columns
data['Club'].isna().values.any()
```

```python
#filling missing value and checking
data['Club'].fillna(data['Club'].value_counts().index[0],inplace=True)
data['Club'].isna().values.any()
```

```python
#checking for missing value in string columns
data['Club Logo'].isna().values.any()
```

```python
#checking for missing value in string columns
data['Preferred Positions'].isna().values.any()
```

```python
#applying for loop for checking and replacing missing values with mean
for col in data:
    if data[col].isna().values.any()==True:
        data[col].fillna(data[col].mean(), inplace=True)
        
#finding if any value is missing
data.isna().values.any()
```

```python
#for confirmation againg checking count of missing values
data.isna().sum().sum()
```

```python
#saving this corrected file in csv format
data.to_csv('FIFA_Project.csv')
```
