```python
import pandas as pd

#taking csv file
fifa_data=pd.read_csv('E:\DBMI\Mini Project\FIFA18CompletePlayerDataset.csv', index_col=0)
fifa_data.describe()
```

```python
from pandas import DataFrame

df = fifa_data[fifa_data['Overall'] > 86] # extracting players with overall above 86
df
```

```python
pd.set_option('display.max_rows', 70)
df.isnull().sum()
```

```python
df = df.fillna(df.mean())
```

```python
df.isnull().sum()
```

```python
names = df['Name'].tolist() # saving names for later

df = df.drop(['Name'], axis = 1) # drop the short_name column 
```

```python
df.head()
```

```python
del df['Photo']
del df['Nationality']
del df['Flag']
del df['Club']
del df['Club Logo']
del df['Preferred Positions']
```

```python
del df['Value']
del df['Wage']
```

```python
#normalize the data
from sklearn import preprocessing

x = df.values # numpy array
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(x)
X_norm = pd.DataFrame(x_scaled)
```

```python
#reducing PCA(Principal Component Analysis) into 2 columns
from sklearn.decomposition import PCA

pca = PCA(n_components = 2) # 2D PCA for the plot
reduced = pd.DataFrame(pca.fit_transform(X_norm))
```

```python
#Performing KMeans Clustering
from sklearn.cluster import KMeans

# specify the number of clusters
kmeans = KMeans(n_clusters=5)

# fit the input data
kmeans = kmeans.fit(reduced)

# get the cluster labels
labels = kmeans.predict(reduced)

# centroid values
centroid = kmeans.cluster_centers_

# cluster values
clusters = kmeans.labels_.tolist()
```

```python
#making new data frame by adding player's names and clusters
reduced['cluster'] = clusters
reduced['name'] = names
reduced.columns = ['x', 'y', 'cluster', 'name']
reduced.head()
```

```python
#Visualization
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set(style="white")

ax = sns.lmplot(x="x", y="y", hue='cluster', data = reduced, legend=False,
                   fit_reg=False, size = 15, scatter_kws={"s": 250})

texts = []
for x, y, s in zip(reduced.x, reduced.y, reduced.name):
    texts.append(plt.text(x, y, s))

ax.set(ylim=(-2, 2))
plt.tick_params(labelsize=15)
plt.xlabel("PC 1", fontsize = 20)
plt.ylabel("PC 2", fontsize = 20)

plt.show()
```

```python
#normalize the data
from sklearn import preprocessing

x = df.values # numpy array
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(x)
X_norm = pd.DataFrame(x_scaled)
```

```python
#reducing PCA(Principal Component Analysis) into 2 columns
from sklearn.decomposition import PCA

pca = PCA(n_components = 2) # 2D PCA for the plot
reduced = pd.DataFrame(pca.fit_transform(X_norm))
```

```python
#Performing KMedoids Clustering
from sklearn_extra.cluster import KMedoids

# specify the number of clusters
kmedoids = KMedoids(n_clusters=5)

# fit the input data
kmeans = kmedoids.fit(reduced)

# get the cluster labels
labels = kmedoids.predict(reduced)

# centroid values
centroid = kmedoids.cluster_centers_

# cluster values
clusters = kmedoids.labels_.tolist()
```

```python
#making new data frame by adding player's names and clusters
reduced['cluster'] = clusters
reduced['name'] = names
reduced.columns = ['x', 'y', 'cluster', 'name']
reduced.head()
```

```python
#Visualization
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set(style="white")

ax = sns.lmplot(x="x", y="y", hue='cluster', data = reduced, legend=False,
                   fit_reg=False, size = 15, scatter_kws={"s": 250})

texts = []
for x, y, s in zip(reduced.x, reduced.y, reduced.name):
    texts.append(plt.text(x, y, s))

ax.set(ylim=(-2, 2))
plt.tick_params(labelsize=15)
plt.xlabel("PC 1", fontsize = 20)
plt.ylabel("PC 2", fontsize = 20)

plt.show()
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch

# plotting dendrogram using average linkage
plt.figure(figsize=(10,14))
plt.title('Hierarchical Clustering Dendrogram with Average Linkage')

dendrogram = sch.dendrogram(sch.linkage(X_norm, method="average"), labels= names, leaf_font_size = 13, orientation='right')
```

```python
# plotting dendrogram using single linkage
plt.figure(figsize=(10,14))
plt.title('Hierarchical Clustering Dendrogram with Single Linkage')

dendrogram = sch.dendrogram(sch.linkage(X_norm, method="single"), labels= names, leaf_font_size = 13, orientation='right')
```

```python
# plotting dendrogram using centroid linkage
plt.figure(figsize=(10,14))
plt.title('Hierarchical Clustering Dendrogram with Centroid Linkage')

dendrogram = sch.dendrogram(sch.linkage(X_norm, method="centroid"), labels= names, leaf_font_size = 13, orientation='right')
```

```python
# plotting dendrogram using complete linkage
plt.figure(figsize=(10,14))
plt.title('Hierarchical Clustering Dendrogram with Complete Linkage')

dendrogram = sch.dendrogram(sch.linkage(X_norm, method="complete"), labels= names, leaf_font_size = 13, orientation='right')
```
