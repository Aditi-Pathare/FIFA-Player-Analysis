```python
# Convertting wages and values to floats
def str_to_float(string):
    if string[-1] == 'M':
        value = float(string[1:-1])*1000000
    elif string[-1] == 'K':
        value = float(string[1:-1])*1000
    else:
        value = float(string[1:])
    return value
 
fifa_data=pd.read_csv('E:\DBMI\Mini Project\FIFA_Project.csv', index_col=0)

#function call
fifa_data['Value']=fifa_data['Value'].apply(str_to_float)
fifa_data['Wage'] = fifa_data['Wage'].apply(str_to_float)

fifa_data.head(8)
```

```python
#taking out the rows on which we want to perform current experiment
df = fifa_data.loc[:,['Name','Age','Nationality','Club','Overall','Potential','Value','Wage']]
df.head(8)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# taking values in variables
height = [0.8505101958934316, 0.5190622195147234, 0.15000570798237148, -0.21826428118960467]
bars = ('Value & Wage', 'Potential & Wage', 'Age & Wage', 'Age & Potential')
x_pos = np.arange(len(bars))
 
# Creating Points
plt.plot(x_pos, height, marker='o', color='g')
 
# Adding the title and y axis name
plt.title('Comparision in Correlations')
plt.ylabel('VALUES')
 
# Creating names needed on the x axis
plt.xticks(x_pos, bars)
 
# Showing graph
plt.show()
```

```python
#Linear Regression
from scipy import stats

#function for plotting
def myfunc(x):
  return slope * x + intercept

#taking columns in variable
x = pd.Series(df['Wage'])
y = pd.Series(df['Age'])
z = pd.Series(df['Potential'])
a = pd.Series(df['Value'])

#graph 1 - Wage vs Age
slope, intercept, r, p, std_err = stats.linregress(x, y)

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.xlabel('WAGE')
plt.ylabel('AGE')
plt.show()

#graph 2 - Wage vs Potential
slope, intercept, r, p, std_err = stats.linregress(x, z)

mymodel = list(map(myfunc, x))

plt.scatter(x, z)
plt.plot(x, mymodel)
plt.xlabel('WAGE')
plt.ylabel('POTENTIAL')
plt.show()

#graph 3 - Potential vs Age
slope, intercept, r, p, std_err = stats.linregress(z, y)

mymodel = list(map(myfunc, z))

plt.scatter(z, y)
plt.plot(z, mymodel)
plt.xlabel('POTENTIAL')
plt.ylabel('AGE')
plt.show()

#graph 4 - Wage vs Value
slope, intercept, r, p, std_err = stats.linregress(x, a)

mymodel = list(map(myfunc, x))

plt.scatter(x, a)
plt.plot(x, mymodel)
plt.xlabel('WAGE')
plt.ylabel('VALUE')
plt.show()

#1e8 = 10^8
```
